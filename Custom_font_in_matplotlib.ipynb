# Custom font in matplotlib

As suggested by Korakot Chaovavanich in https://stackoverflow.com/a/51844978/1840471 ([Gist](https://gist.github.com/korakot/9d7f5db632351dc92607fdec72a4953f)).

```
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```
!wget https://github.com/MaxGhenis/random/raw/master/Roboto-Regular.ttf -P /usr/local/lib/python3.6/dist-packages/matplotlib/mpl-data/fonts/ttf
```

```
mpl.font_manager._rebuild()
```

Example simplified from https://matplotlib.org/examples/pylab_examples/simple_plot.html. Show first with the default, then with the new font.

```
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2*np.pi*t)
```

```
plt.plot(t, s)

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.show()
```

```
mpl.rc('font', family='Roboto')
```

```
plt.plot(t, s)

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.show()
```
