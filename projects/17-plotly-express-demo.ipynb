```
import plotly_express as px
import warnings
warnings.filterwarnings("ignore")
```

```
# A built-in dataset
tips = px.data.tips()
tips.head()
```

The trendlines are of approximately the same slope, so men and women tip roughly the same.

```
px.scatter(tips, x='total_bill', y='tip', marginal_x='box', marginal_y='violin', color='sex', trendline='ols')
```

```
# We can quickly order by a specific category by throwing it all in a dictionary
px.scatter(tips, x='total_bill', y='tip', facet_col='day', facet_row='time',
           category_orders={'day':['Thur', 'Fri', 'Sat', 'Sun']})
```

```
px.bar(tips, x='day', y='total_bill', category_orders={'day':['Thur', 'Fri', 'Sat', 'Sun']})
```

```
# Summary plotting
px.scatter_matrix(tips)
```
